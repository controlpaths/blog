```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
nSamples = 128;
angle = np.linspace(0,2*np.pi,nSamples, endpoint=False)
sineSignal = np.sin(angle)*0.9
plt.stem(sineSignal, use_line_collection='true')
```

```python
nBits = 24
signalQuantified = (sineSignal*2**(nBits-1))+2**(nBits-1)
```

```python
# creating the file 
f = open("sine24bits_128samples.mem", "w")
for i in signalQuantified:
    f.write(format(int(i),'x') + ' ')

f.close()
```
