```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
fs = 96e3;
fc = 1e3;

fNyquist = fs/2;
wc = fc/fNyquist;

nTaps = 32;

taps = signal.firwin(nTaps, wc)

plt.stem(taps, use_line_collection='true')
```

```python
w,h = signal.freqz(taps)
plt.plot(w/np.pi*fNyquist, 20 * np.log10(abs(h)), 'b')
plt.grid()
print(20*np.log10(abs(h[50])))
```

```python
# creating the file 
f = open("fir.coe", "w")
f.write("radix=10;\n")
f.write("coefdata=\n")

for i in taps:
    f.write(str(i) + '\n')

f.write(';')
f.close()
```
